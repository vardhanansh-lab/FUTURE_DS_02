```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("telco_churn.xlsx")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df['Churn'].value_counts()
```

```python
sns.countplot(x='Churn', data=df)
plt.title("Customer Churn Distribution")
plt.show()
```

```python
sns.countplot(x='Contract', hue='Churn', data=df)
plt.title("Churn by Contract Type")
plt.show()
```

```python
sns.boxplot(x='Churn', y='tenure', data=df)
plt.title("Tenure vs Churn")
plt.show()
```

```python
sns.boxplot(x='Churn', y='MonthlyCharges', data=df)
plt.title("Monthly Charges vs Churn")
plt.show()
```

# Key Insights

Customers with month-to-month contracts churn the most

Customers with higher tenure stay longer

Higher monthly charges slightly increase churn

# Recommendations

Encourage long-term contracts with discounts

Offer loyalty rewards after 12 months

Provide retention offers for high-charge customers

```python
sns.countplot(x='PaymentMethod', hue='Churn', data=df)
plt.title("Churn by Payment Method")
plt.xticks(rotation=45)
plt.show()
```

Customers paying via Electronic Check show higher churn.

Automatic payment methods have better retention.

```python
sns.countplot(x='InternetService', hue='Churn', data=df)
plt.title("Churn by Internet Service")
plt.show()
```

Customers using Fiber Optic service show higher churn.

DSL users appear more stable.

```python
sns.countplot(x='gender', hue='Churn', data=df)
plt.title("Churn by Gender")
plt.show()
```

Churn distribution across gender is relatively balanced.

```python
sns.histplot(df['tenure'], bins=30)
plt.title("Customer Tenure Distribution")
plt.show()
```

Most customers have low tenure, indicating potential retention challenges.

# **Key Retention Drivers**

Contract type strongly influences churn.

Customers with month-to-month contracts churn more frequently.

Customers with longer tenure show stronger loyalty.

Electronic check payments correlate with higher churn.

# **Business Recommendations**

Encourage customers to shift to long-term contracts.

Provide discount incentives for yearly subscriptions.

Improve customer experience for Fiber Optic users.

```python
churn_rate = df['Churn'].value_counts(normalize=True) * 100
print(churn_rate)
```

```python
df['Churn'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    colors=['lightblue','salmon']
)

plt.title("Customer Retention vs Churn")
plt.ylabel("")
plt.show()
```

# **Overall Retention Analysis**

• Approximately 73% of customers are retained.

• Around 27% customers churn from the service.

This indicates that while the company maintains a strong retention base, churn drivers such as contract type and payment methods should be optimized to improve retention further.
